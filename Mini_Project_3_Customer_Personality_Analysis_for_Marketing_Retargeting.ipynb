# Feature Engineering

## Conversion Rate

```python
# define conversion rate feature
def conversion_rate(a,b):
    if b == 0:
        conversion_rate = 0
    else:
        conversion_rate = a/b
    return conversion_rate
```

```python
# feature engineering: conversion rate
df['conversion_rate'] = df.apply(lambda x : conversion_rate(x['Response'], x['NumWebVisitsMonth']), axis = 1)
df['conversion_rate'] = round(df['conversion_rate'], 2)
df.info()
```

## Customers Age

```python
# feature engineering: age
df['age'] = pd.Timestamp('now').year - df['Year_Birth']
df.head()
```

## Customers Age Group

```python
# kdeplot from age feature
sns.kdeplot(data = df, x = 'age')
plt.show()
```

```python
# print min and max age in dataset
print('minimum umur adalah', min(df['age']))
print('maximum umur adalah', max(df['age']))
```

**Pembagian fitur age group berdasarkan jurnal <a href="https://academic.oup.com/gerontologist/article/42/1/92/641498">ini</a>**
- young adults (ages 18â€“35 years)
- middle-aged adults (ages 36â€“55 years)
- older adults (aged older than 55 years)

```python
# define age group feature
def age_group(x):
    if x['age'] >= 18 and x['age']<= 35:
        age_group = 'young adults'
    elif x['age'] >= 36 and x['age'] <= 55:
        age_group = 'middle-aged adults'
    else:
        age_group = 'older adults'
    return age_group
```

```python
# feature engineering: age group
df['age_group'] = df.apply(lambda x : age_group(x), axis = 1)
df.sample(5)
```

```python
df.info()
```

## Total Accepted Campaign

```python
# feature engineering: total accepted campaign
df['total_acc_camp'] = df['AcceptedCmp1'] + df['AcceptedCmp2'] + df['AcceptedCmp3']
+ df['AcceptedCmp4'] + df['AcceptedCmp5']
df.head()
```

```python
df.columns
```

## Total Spent

```python
# feature engineering: total spent
df['total_spent'] = df['MntCoke'] + df['MntFruits'] + df['MntMeatProducts']
+ df['MntFishProducts'] + df['MntSweetProducts'] + df['MntGoldProds']
df.info()
```

## Total Purchase

```python
# feature engineering: total purchase
df['total_purchase'] = df['NumDealsPurchases'] + df['NumWebPurchases'] + df['NumCatalogPurchases']
+ df['NumStorePurchases']
df.head()
```

## Number of Children

```python
# feature engineering: number of children
df['numb_children'] = df['Kidhome'] + df['Teenhome']
df.head()
```

## Total Days Joined

```python
df.info()
```

```python
# change Dt_Customer data type from object to datetime
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# feature engineering: total_days_joined
df['total_days_joined'] = (pd.Timestamp('now') - df['Dt_Customer']).dt.days

# feature engineering : years_customer
df['years_customer'] = (pd.Timestamp('now').year) - (df['Dt_Customer']).dt.year
df.head()
```

Berdasarkan dua jurnal referensi berikut mengenai segmentation di marketing:
- Jurnal 1, klik <a href="https://www.researchgate.net/publication/326771853_A_New_Approach_for_Customer_Clustering_by_Integrating_the_LRFM_Model_and_Fuzzy_Inference_System">disini</a>

Ada 4 komponen, yaitu:
1. Length (L): The number of days from the first to the last visit date in a given time period
2. Recency (R): The number of days since the last purchase in a given time period
3. Frequency (F): The number of purchase made in a given time period
4. Monetary (M): The total amount of money spent during a given period of time

Kolom yang akan dipilih berdasarkan LRFM Analysis tersebut:
- L (Loyalty/Length): lama hari customer sejak menjadi member pertama kali
- R (Recency): jumlah hari sejak pembelian terakhir customer
- F (Frequency): total jumlah pembelian yang dilakukan
- M (Monetary): jumlah yang dihabiskan untuk keseluruhan produk

sehingga,
- L : 'total_days_joined'
- R : 'Recency'
- F : 'total_purchase'
- M : 'total_spent'

## Handling Outlier

```python
# outlier removal with IQR
print("Jumlah sebelum filter: ", len(df_fix))

for i in num_fix:
  Q1 = df_fix[i].quantile(0.25)
  Q3 = df_fix[i].quantile(0.75)
  IQR = Q3-Q1
  low_limit = Q1 - (1.5 * IQR)
  high_limit = Q3 + (1.5 * IQR)
  filtered_entries = ((df_fix[i] >= low_limit) & (df_fix[i] <= high_limit))
  df_fix = df_fix[filtered_entries]

print("Jumlah setelah filter:", len(df_fix))
```

## Elbow Method

```python
from sklearn.cluster import KMeans
inertia = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 42)
    kmeans.fit(df_std.values)
    inertia.append(kmeans.inertia_)
```

```python
sns.lineplot(x = range(1,11), y = inertia, linewidth = 4)
sns.scatterplot(x = range(1,11), y = inertia, s = 60)
plt.tight_layout()
```

```python
pd.Series(inertia) - pd.Series(inertia).shift(-1)
```

Berdasarkan elbow method, jumlah cluster efisien di 3 cluster, yaitu k atau n_clusters = 3. Akan tetapi, dari visualisasinya belum meyakinkan soal pemilihan clusternya. Oleh sebab itu, akan dilakukan visualisasi distortion score dan shilouette score elbow untuk menentukan jumlah cluster yang tepat.

## Visualization of Distortion Score and Silhouette Score Elbow

```python
# Distortion Score Elbow
from yellowbrick.cluster import KElbowVisualizer

# fit model
model = KMeans(random_state=42)
visualizer = KElbowVisualizer(model, k = (2,11), metric = 'distortion', timings = True, locate_elbow = True)
visualizer.fit(df_std)
visualizer.show()
```

```python
# silhouette score elbow
model = KMeans(random_state = 42)
visualizer = KElbowVisualizer(model, k = (2,11), metric = 'silhouette', timings = True, locate_elbow = True)
visualizer.fit(df_std)
visualizer.show()
```

Pada visualisasi distortion score dan shilouette score, nilai k atau n_cluster yang diperoleh berbeda, sehingga kita akan membandingkan nilai average shilouette score pada range 2 sampai 6.

## Silhouette Plot and Silhouette Score

```python
# silhouette plot
from yellowbrick.cluster import SilhouetteVisualizer

for i in range(2,7):
    model = KMeans(i, random_state = 42)
    visualizer = SilhouetteVisualizer(model, colors = 'yellowbrick')
    visualizer.fit(df_std)
    visualizer.show()
```

```python
from sklearn.metrics import silhouette_score
for num_clusters in range(2,7):
    #inisialisasi kmeans
    model_clus = KMeans(n_clusters = num_clusters, max_iter = 1000, random_state = 42)
    model_clus.fit(df_std)

    cluster_labels = model_clus.labels_

    #shilhouette score
    silhouette_avg = silhouette_score(df_std, cluster_labels)
    print("For n_clusters = {0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

Dari hasil output di atas, jumlah cluster yang memiliki komposisi yang seimbang dengan setiap anggota cluster dan memperhatikan average shilouette score dari masing-masing cluster adalah 3 cluster. Walaupun 6 cluster memiliki average shilouette score yang lebih tinggi dibandingkan 3 cluster, 6 cluster memiliki komposisi tiap anggota cluster yang tidak seimbang dibandingkan dengan 3 cluster, begitu juga untuk 2 cluster memiliki komposisi tiap anggota cluster yang tidak seimbang meskipun silhoutte scorenya lebih tinggi.
